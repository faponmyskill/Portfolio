## Предобработка данных

Предобработку данных начнем с детального анализа столбцов с пустыми значениями и их заменой на корректные. Поскольку нам местоположение - одна из основных значений данной таблицы и сделать предположений по точной геолокации мы не можем, то произведем удаление строчек с пустым значением местоположения.

```python
estate_data.dropna(subset = ['locality_name'], inplace = True)
estate_data.info()
```

Заменим некорректные названия посёлков, сделав их уникальными.

### Подсчет и добавление этаж квартиры; варианты — первый, последний, другой


floor — этаж
floors_total — всего этажей в доме

```python
def floor_type(row):
    floor = row['floor'] #этаж записываем в floor 
    floors_total = row['floors_total'] #записываем какой последний этаж у дома
    if floor == 1:
        return 'первый этаж'
    if floor == floors_total:
        return 'последний этаж'
    return 'другой'
    
data_final['floor_type'] = data_final.apply(floor_type, axis = 1)
data_final['floor_type'].unique()
```
