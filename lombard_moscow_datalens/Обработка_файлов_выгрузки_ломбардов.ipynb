```python
# Функция для объединения координат
def combine_coordinates(row):
    return [row['Shirota'], row['Dolgota']]

# Создаем новый столбец 'Coordinates'
df_metro['Coordinates'] = df_metro.apply(combine_coordinates, axis=1)
```

```python
from geopy.distance import geodesic

# Функция для создания контура
def create_contour(coords, radius):
    nw = geodesic(kilometers=radius).destination(coords, 315)
    ne = geodesic(kilometers=radius).destination(coords, 45)
    se = geodesic(kilometers=radius).destination(coords, 135)
    sw = geodesic(kilometers=radius).destination(coords, 225)
    return [[nw.longitude, nw.latitude], [ne.longitude, ne.latitude], [se.longitude, se.latitude], [sw.longitude, sw.latitude]]

# Функция для создания полигона с "дыркой"
def create_polygon(coords, outer_radius=1, inner_radius=0.1):
    outer_contour = create_contour(coords, outer_radius)
    inner_contour = create_contour(coords, inner_radius)
    return [outer_contour, inner_contour]

# Применяем функцию к каждому ряду в DataFrame
df_metro['Polygon'] = df_metro['Coordinates'].apply(create_polygon)
```

```python
from geopy.distance import geodesic

# Функция для создания контура
def create_contour(coords, radius):
    points = []
    for degree in range(0, 360, 10):
        point = geodesic(kilometers=radius).destination(coords, degree)
        points.append([point.longitude, point.latitude])
    return points

# Функция для создания полигона с "дыркой"
def create_polygon(coords, outer_radius=1, inner_radius=0.1):
    outer_contour = create_contour(coords, outer_radius)
    inner_contour = create_contour(coords, inner_radius)
    return [outer_contour, inner_contour]

# Применяем функцию к каждому ряду в DataFrame
df_metro['Polygon'] = df_metro['Coordinates'].apply(create_polygon)
```

```python
import pandas as pd
import pyproj
from shapely.geometry import Point, Polygon
from shapely.ops import transform
from functools import partial

# функция для создания геометрии полигона вокруг заданной точки
def buffer_in_meters(lon, lat, radius):
    proj_wgs84 = pyproj.Proj('+proj=longlat +datum=WGS84')

    # азимутальная эквидистантная проекция для точки
    aeqd_proj = '+proj=aeqd +lat_0={lat} +lon_0={lon} +x_0=0 +y_0=0'
    project = partial(
        pyproj.transform,
        pyproj.Proj(aeqd_proj.format(lat=lat, lon=lon)),
        proj_wgs84)

    buf = Point(0, 0).buffer(radius * 1000)  # radius in kilometers

    return transform(project, buf).exterior.coords[:]

# функция для создания геополигона с "дыркой"
def create_polygon(coords, outer_radius=1, inner_radius=0.012):
    outer_contour = buffer_in_meters(coords[1], coords[0], outer_radius)
    inner_contour = buffer_in_meters(coords[1], coords[0], inner_radius)
    return [list(map(list, outer_contour)), list(map(list, inner_contour))]



# применяем функцию к каждому ряду в DataFrame
df_metro['Polygon'] = df_metro['Coordinates'].apply(create_polygon)
```
