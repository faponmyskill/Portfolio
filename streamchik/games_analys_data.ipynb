# Выявление закономерностей успешности игры

## Описание проекта

В проекте представлены данные интернет - магазина «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Необходимо выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

**Описание данных**
<br><br>
<br>Name — название игры
<br>Platform — платформа
<br>Year_of_Release — год выпуска
<br>Genre — жанр игры
<br>NA_sales — продажи в Северной Америке (миллионы проданных копий)
<br>EU_sales — продажи в Европе (миллионы проданных копий)
<br>JP_sales — продажи в Японии (миллионы проданных копий)
<br>Other_sales — продажи в других странах (миллионы проданных копий)
<br>Critic_Score — оценка критиков (максимум 100)
<br>User_Score — оценка пользователей (максимум 10)
<br>Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

## Проверка гипотез

Проверим несколько гипотез. Зададим пороговое значение уровня статистической значимости alpha = 0.05, поскольку нам не требуется более точное определение. 

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;


Проведем сравнение двух совокупностей и проверим нулевую гипотезу: "Средние пользовательские рейтинги платформ Xbox One и PC одинаковые". Тогда альтернативная: "Средние пользовательские рейтинги платформ Xbox One и PC не одинаковые". И по скольку у нас есть исходные данные двух сравниваемых массивов, установим параметр equal_var = False под расчет дисперсий.

```python
from scipy import stats as st
```

```python
users_rating_xbox = data['user_score'][(data['platform'] == "XOne") & (data['user_score'].notnull())]
users_rating_pc = data['user_score'][(data['platform'] == "PC")  & (data['user_score'].notnull())]


alpha = 0.05 # уровень статистической значимости

results = st.ttest_ind(users_rating_xbox, users_rating_pc, equal_var = False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Исходя из теста можем предположить, что гипотеза "Средние пользовательские рейтинги платформ Xbox One и PC одинаковые" - верна, однако, полученное значение p-value говорит о том, что с вероятностью в 11% такое или большее различие можно получить случайно. Вероятность не сильно высокая, а значит люди, скорее всего, одинакого хорошо реагируют на платформу Xbox One и PC.

### Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Проведем сравнение двух совокупностей и проверим нулевую гипотезу: "Средние пользовательские рейтинги жанров Action и Sports равны". Тогда альтернативная: "Средние пользовательские рейтинги жанров Action и Sports не равны, а значит - разные". И по скольку у нас есть исходные данные двух сравниваемых массивов, установим параметр equal_var = False под расчет дисперсий.

```python
users_rating_action = data['user_score'][(data['genre'] == "Action") & (data['user_score'].notnull())]
users_rating_sports = data['user_score'][(data['genre'] == "Sports")  & (data['user_score'].notnull())]


alpha = 0.05 # уровень статистической значимости

results = st.ttest_ind(users_rating_action, users_rating_sports, equal_var = False)
print('Среднее по Actions', users_rating_action.mean())
print('Среднее по Sports', users_rating_sports.mean())

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Полученное значение p-value говорит о том, что средние пользовательские рейтинги жанров Action и Sports, с крайне низкой вероятностью могут быть равны между собой.
